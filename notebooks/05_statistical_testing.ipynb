## 📈 Skewness Analysis

Skewness measures the **asymmetry of a distribution**.

- **Positive skew** → long tail on the right
- **Negative skew** → long tail on the left
- **Zero skew** → symmetric distribution

Understanding skewness helps determine whether transformations (such as log scaling) are needed to stabilize the data distribution.

```python
for col in pd_df.columns:
    skew = pd_df[col].skew()
    print(f"{col} Skewness: {skew:.5f}")
    if skew == 0:
        print(f"{col} is symmetric.")
    elif skew < 0:
        print(f"{col} is negatively skewed.")
    else:
        print(f"{col} is positively skewed.")
    print()
```

## 🔄 Log Transformation

Financial and transactional datasets often exhibit **right-skewed distributions**.

To reduce skewness and improve statistical properties, we apply a **log(1 + x) transformation** to numerical variables.

This transformation helps:

- compress extreme values
- stabilize variance
- improve model performance

```python
log_df = pd_df.copy()

for col in pd_df.columns:
    log_df[col] = np.log1p(pd_df[col])
```

## 🚫 Outlier Treatment (IQR Method)

Extreme values can distort statistical analysis and machine learning models.

We apply the **Interquartile Range (IQR) method** to cap extreme observations.

Formula:

Lower Bound = Q1 − 1.5 × IQR  
Upper Bound = Q3 + 1.5 × IQR

Values outside these limits are capped to reduce their influence while preserving data structure.

```python
clean_df = log_df.copy()

for col in clean_df.columns:
    Q1 = clean_df[col].quantile(0.25)
    Q3 = clean_df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    clean_df[col] = np.where(clean_df[col] < lower, lower,
                     np.where(clean_df[col] > upper, upper,
                              clean_df[col]))
```

### Distribution After Transformation & Outlier Treatment

After applying log transformation and outlier capping, we re-examine the distributions of numerical features.

The goal is to verify whether the transformations successfully reduced skewness and extreme values.

```python
sns.set_style("whitegrid")

numeric_cols = clean_df.columns

plt.figure(figsize=(15, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 2, i)
    sns.histplot(clean_df[col], kde=True,kde_kws={'bw_adjust': 2},color="#14B8A6")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## 📉 Kurtosis Analysis

Kurtosis measures the **tailedness of a distribution**.

- **Leptokurtic (>0)** → heavy tails and extreme outliers
- **Mesokurtic (=0)** → normal distribution
- **Platykurtic (<0)** → light tails

Understanding kurtosis helps assess the likelihood of extreme observations.

```python
for col in pd_df.columns:
    kurtosis = pd_df[col].kurtosis()
    print(f"{col} kurtosis: {kurtosis:.5f}")
    if kurtosis == 0:
        print(f"{col} is mesokurtic.")
    elif kurtosis < 0:
        print(f"{col} is platykurtic.")
    else:
        print(f"{col} is leptokurtic.")
    print()
```

```python
kurtosis_comparison = pd.DataFrame({
    "Original": pd_df.kurtosis(),
    "Log_Transformed": log_df.kurtosis(),
    "After_Outlier_Treatment": clean_df.kurtosis()
})

kurtosis_comparison
```

```python
kurtosis_comparison.plot(
    kind="bar", 
    figsize=(12,6),
    color = ["#14B8A6", "#6B7280", "#3B82F6"]
)
plt.title("Kurtosis Comparison")
plt.xticks(rotation=0)
plt.show()
```

## 🧪 Normality Testing

To statistically evaluate distribution normality, we use the **Shapiro–Wilk test**.

- **H₀:** Data follows a normal distribution  
- **H₁:** Data does not follow a normal distribution

However, in large datasets, even small deviations from normality often produce significant p-values. Therefore, statistical tests are interpreted alongside visual inspections.

```python
for col in clean_df.columns:
    stat, p = stats.shapiro(clean_df[col].sample(5000, random_state=42))
    print(f"{col} p-value = {p:.4f}")
    if p > 0.05:
        print(f"{col} is normally distributed.")
    else:
        print(f"{col} is not normally distributed.")
    print()
```

## ⚖️ Feature Scaling

Machine learning algorithms often perform better when numerical features are on similar scales.

We apply **Standard Scaling**, which transforms features using:

z = (x − mean) / standard deviation

This ensures each feature has:

- mean ≈ 0
- standard deviation ≈ 1


```python
scaler = StandardScaler()

scaled_features = scaler.fit_transform(clean_df)

scaled_df = pd.DataFrame(
    scaled_features,
    columns=clean_df.columns
)

scaled_df.head()
```

## 🔗 Multicollinearity Analysis (VIF)

Multicollinearity occurs when independent variables are highly correlated.

We measure multicollinearity using the **Variance Inflation Factor (VIF)**.

Interpretation:

- **VIF < 5** → low multicollinearity  
- **5 ≤ VIF < 10** → moderate multicollinearity  
- **VIF ≥ 10** → severe multicollinearity

```python
X = scaled_df.copy()

for i in range(X.shape[1]):
    
    vif_value = variance_inflation_factor(X.values, i)
    feature_name = X.columns[i]
    
    print(f"{feature_name} : VIF = {vif_value:.4f}")
    
    if vif_value > 10:
        print(f"🚨 {feature_name} has Severe multicollinearity")
        print(f"{feature_name} is a strong candidate for removal")
        
    elif vif_value > 5:
        print(f"⚠️ {feature_name} has High multicollinearity")
        print(f"{feature_name} may be considered for removal")
        
    else:
        print(f"✅ {feature_name} has Low multicollinearity")
        print(f"{feature_name} is safe to retain")
    
    print("-" * 50)
```

# 📊 Statistical Hypothesis Testing

To validate insights observed during exploratory analysis, we perform statistical hypothesis tests.

These tests help determine whether observed differences in the dataset are **statistically significant or due to random variation**.

Tests performed:

1️⃣ Discount impact on order value (t-test)  
2️⃣ Category impact on revenue (ANOVA)  
3️⃣ Payment method impact on spending (ANOVA)

## 🏷️ Hypothesis 1: Do Discounts Significantly Impact Order Value?

**Null Hypothesis (H₀):** There is no difference in average order value between discounted and non-discounted orders.

**Alternative Hypothesis (H₁):** Discounted orders have a significantly different average order value.

```python
pdf = df.select(
    "TotalAmount",
    f.when(f.col("Discount") > 0, 1).otherwise(0).alias("IsDiscounted")
).toPandas()

discounted = pdf[pdf["IsDiscounted"] == 1]["TotalAmount"]
non_discounted = pdf[pdf["IsDiscounted"] == 0]["TotalAmount"]

t_stat, p_value = stats.ttest_ind(discounted, non_discounted, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

alpha = 0.05 

if p_value < alpha:
    print("→ We reject the null hypothesis.")
    print("→ Discounts significantly impact order value.")
else:
    print("→ We fail to reject the null hypothesis.")
    print("→ No statistically significant impact detected.")
```

## 📌 Hypothesis Test Insight: Impact of Discounts on Order Value

A two-sample t-test (Welch’s t-test) was conducted to determine whether discounted orders significantly differ in average order value compared to non-discounted orders.

### 🧪 Test Results:
- **T-Statistic:** -0.637  
- **P-Value:** 0.524  
- **Significance Level (α):** 0.05  

### 📊 Interpretation:

Since the p-value (0.524) is significantly greater than the chosen significance level of 0.05, we **fail to reject the null hypothesis**.

This indicates that there is **no statistically significant difference** in average order value between discounted and non-discounted orders.

### 💡 Business Insight:

Although discounts are commonly used to drive higher sales volumes, the statistical results suggest that discounts do not meaningfully impact the overall order value in this dataset.

This may indicate:

- Customers purchase similar-value baskets regardless of discounts.
- Discounts may influence purchase decision timing rather than order size.
- Pricing strategy may not be strongly affecting revenue per transaction.

### 🎯 Recommendation:

Further analysis could explore:
- Whether discounts impact purchase frequency instead of order value.
- If high-discount segments behave differently from low-discount segments.
- The long-term effect of discounts on customer retention and lifetime value.

## 🛍️ Hypothesis 2: Do Different Product Categories Generate Different Average Revenue?

**H₀:** All categories have the same average order value.  
**H₁:** At least one category has a different average order value.

```python
pdf = df.select("Category", "TotalAmount", "Discount", "PaymentMethod").toPandas()

category_groups = [
    group["TotalAmount"].values
    for name, group in pdf.groupby("Category")
]

anova_result = stats.f_oneway(*category_groups)

print("F-Statistic:", anova_result.statistic)
print("P-Value:", anova_result.pvalue)

if anova_result.pvalue < alpha:
    print("→ We reject the null hypothesis.")
    print("→ At least one category has a significantly different average order value.")
else:
    print("→ We fail to reject the null hypothesis.")
    print("→ No statistically significant difference detected.")
```

### 📊 ANOVA Results Interpretation

The one-way ANOVA test was conducted to determine whether average order values differ significantly across product categories.

**Results:**
- F-Statistic: 1.48  
- P-Value: 0.192  

### 📌 Conclusion

Since the p-value (0.192) is greater than the significance level (α = 0.05), we **fail to reject the null hypothesis**.

This indicates that there is **no statistically significant difference in average order value across product categories** in this dataset.

### 💡 Business Interpretation

- Although some categories may visually appear to generate higher revenue, the observed differences are not statistically significant.
- This suggests that pricing and spending behavior may be relatively consistent across categories.
- Strategic decisions such as pricing differentiation should consider additional factors (e.g., volume, customer segments, seasonal effects) rather than category alone.

### 🎯 Analytical Insight

This result highlights the importance of statistical validation. Visual differences in EDA do not always translate into statistically meaningful variations.

## 💳 Hypothesis 3: Does Payment Method Influence Customer Spending?

- **Null Hypothesis (H₀):** The average order value is the same across all payment methods.

- **Alternative Hypothesis (H₁):** At least one payment method has a significantly different average order value.

```python
pdf = df.select("PaymentMethod", "TotalAmount").toPandas()

payment_groups = pdf.groupby("PaymentMethod")["TotalAmount"].apply(list)

anova_result = stats.f_oneway(*payment_groups)

print("F-Statistic:", anova_result.statistic)
print("P-Value:", anova_result.pvalue)

if anova_result.pvalue > alpha:
    print("→ We fail to reject the null hypothesis.")
    print("→ No statistically significant difference detected.")
else:
    print("→ We reject the null hypothesis.")
    print("→ At least one payment method has a significantly different average order value.")
    
```

### 📊 Statistical Test Result — Payment Method Impact

The One-Way ANOVA test produced the following results:

- **F-Statistic:** 0.689  
- **P-Value:** 0.632  

Since the p-value (0.632) is greater than the significance level (α = 0.05), we **fail to reject the null hypothesis**.

### 📌 Conclusion

There is no statistically significant difference in average order value across different payment methods.

### 💼 Business Interpretation

This suggests that customers tend to spend similar amounts regardless of whether they use:

- Debit Card  
- Credit Card  
- Amazon Pay  
- Other available payment options  

From a business perspective:

- Payment method does not appear to influence basket size.
- Promotional strategies tied exclusively to payment methods may not significantly increase revenue.
- Revenue optimization efforts may be better focused on pricing strategy, product bundling, or targeted customer segmentation.

### 🧠 Analytical Insight

While descriptive statistics may show small variations in average order value between payment methods, these differences are not statistically significant and could be attributed to random variation rather than meaningful behavioral differences.
