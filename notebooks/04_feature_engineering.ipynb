## 👥 RFM Analysis (Recency, Frequency, Monetary)

RFM analysis is a widely used customer segmentation technique that evaluates
customer value based on purchasing behavior.

The three RFM components are:

- **Recency (R)** – Number of days since the customer's last purchase
- **Frequency (F)** – Total number of orders placed by the customer
- **Monetary (M)** – Total amount spent by the customer

These metrics help identify valuable customers and provide insight into customer
engagement and spending patterns.

The reference date is dynamically determined using the **most recent order date
in the dataset**, ensuring that recency is calculated relative to the dataset timeline.

### RFM Feature Computation

We compute the following values for each customer:

1. **Last Purchase Date**
2. **Recency** – Days since the last purchase
3. **Frequency** – Total orders placed
4. **Monetary Value** – Total spending

These metrics form the foundation for customer segmentation.

```python
reference_date = df.select(f.max("OrderDate")).first()[0]

last_purchase = df.groupBy("CustomerID") \
    .agg(f.max("OrderDate").alias("LastPurchaseDate"))

rfm = customer_metrics.join(last_purchase, "CustomerID") \
    .withColumn("Recency", f.datediff(f.lit(reference_date), "LastPurchaseDate"))

rfm = rfm.select(
    "CustomerID",
    f.col("Recency"),
    f.col("TotalOrders").alias("Frequency"),
    f.col("TotalSpend").alias("Monetary"),   
    "LastPurchaseDate","AvgOrderValue"
)

display(rfm.limit(10))
```

### RFM Feature Distributions

To better understand the distribution of RFM metrics, we visualize both
the raw and log-transformed values.

Customer spending and purchasing frequency often follow **right-skewed
distributions**, where a small number of customers generate a large share
of revenue.

Log transformations help normalize these distributions and make them
more suitable for machine learning models.

```python
rfm_pd = rfm.toPandas()

rfm_pd["RecencyLog"] = np.log1p(rfm_pd["Recency"])
rfm_pd["FrequencyLog"] = np.log1p(rfm_pd["Frequency"])
rfm_pd["MonetaryLog"] = np.log1p(rfm_pd["Monetary"])

plt.figure(figsize=(15,8))

# --- Raw Distributions ---
plt.subplot(2,3,1)
sns.histplot(rfm_pd["Recency"], bins=30, kde=True, color="#2563EB")
plt.title("Recency Distribution")

plt.subplot(2,3,2)
sns.histplot(rfm_pd["Frequency"], bins=30, kde=True, color="#2563EB")
plt.title("Frequency Distribution")

plt.subplot(2,3,3)
sns.histplot(rfm_pd["Monetary"], bins=30, kde=True, color="#2563EB")
plt.title("Monetary Distribution")

# --- Log Transformed ---
plt.subplot(2,3,4)
sns.histplot(rfm_pd["RecencyLog"], bins=30, kde=True, color="#14B8A6")
plt.title("Recency (Log) Distribution")

plt.subplot(2,3,5)
sns.histplot(rfm_pd["FrequencyLog"], bins=30, kde=True,color="#14B8A6" )
plt.title("Frequency (Log) Distribution")

plt.subplot(2,3,6)
sns.histplot(rfm_pd["MonetaryLog"], bins=30, kde=True, color="#14B8A6")
plt.title("Monetary (Log) Distribution")

plt.tight_layout()
plt.show()
```

## 📅 Temporal Feature Engineering

Customer purchasing behavior often varies across different days of the week.

We create two additional features:

- **DayOfWeek** – Day name extracted from the order date
- **IsWeekend** – Boolean indicator showing whether the order occurred on a weekend

These features help capture **weekly purchasing patterns** that may influence
customer behavior.

```python
df_features = df.withColumn(
                    "DayOfweek",
                    f.date_format("OrderDate", "EEEE") 
                ).withColumn(
                    "IsWeekend",
                    f.when(
                        f.col("DayOfweek").isin("Saturday", "Sunday"),
                        True
                    ).otherwise(False)
                )

display(df_features.limit(10))
```

## 🏆 Customer Value Segmentation

Customers are segmented into value tiers based on their **Monetary spending**.

We use quartile-based segmentation:

| Tier| Customer Segment  |
|-----|-------------------|
| 1   | Low Value         |
| 2   | Medium Value      |
| 3   | High Value        |
| 4   | Premium Customers |

This segmentation helps identify high-value customers who contribute
significantly to overall revenue.

```python
quantiles = rfm.approxQuantile("Monetary", [0.25, 0.5, 0.75], 0.01)

q1, q2, q3 = quantiles

customer_rfm = rfm.withColumn(
    "CustomerValueTier",
    f.when(f.col("Monetary") <= q1, 1)
     .when(f.col("Monetary") <= q2, 2)
     .when(f.col("Monetary") <= q3, 3)
     .otherwise(4)
)

display(customer_rfm.limit(10))
```

## 🤖 Machine Learning Feature Dataset

Finally, we combine all engineered features into a single dataset
that will be used for machine learning models.

The resulting dataset includes:

- Customer behavior features (RFM)
- Customer value tier
- Temporal purchasing features
- Order-level attributes

This dataset will serve as the input for downstream modeling tasks
such as customer segmentation or purchase prediction.

```python
df_ml = df_features.join(customer_rfm.select("CustomerID",
                                    "Frequency",
                                    "Monetary",
                                    "AvgOrderValue",
                                    "Recency",
                                    "CustomerValueTier"),
                on="CustomerID",
                how="left")

display(df_ml.limit(10))
```
