# 🤖 Machine Learning: Customer Segmentation (K-Means)

This notebook applies unsupervised machine learning to segment customers using RFM features (Recency, Frequency, Monetary).

Customer segmentation helps businesses understand customer behavior and design targeted marketing strategies.

The dataset used here contains customer-level RFM metrics generated in the Feature Engineering notebook.

## Select RFM Features

For customer segmentation we use the RFM framework:

Recency → How recently a customer made a purchase  
Frequency → How often the customer purchases  
Monetary → Total amount spent

```python
rfm_features = customer_rfm.select(
    "CustomerID",
    "Recency",
    "Frequency",
    "Monetary"
)

display(rfm_features.limit(10))
```

## Convert to Pandas for Machine Learning

Spark DataFrames are converted to Pandas since Scikit-Learn models operate on Pandas/Numpy structures.

```python
rfm_pd = rfm_features.toPandas()

rfm_pd.set_index("CustomerID", inplace=True)

rfm_pd.head()
```

## Feature Scaling

Since RFM variables have different scales, we apply standardization using StandardScaler before clustering.

```python
rfm_scaled = scaler.fit_transform(rfm_pd)

rfm_scaled_df = pd.DataFrame(
    rfm_scaled,
    columns=rfm_pd.columns,
    index=rfm_pd.index
)

rfm_scaled_df.head()
```

## Determining Optimal Number of Clusters

The elbow method helps determine the optimal number of clusters by analyzing the within-cluster sum of squares (inertia).

```python
inertia = []

k_range = range(1,11)

for k in k_range:
    
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled_df)
    
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(16,9))

plt.plot(k_range, inertia, marker="o", color="#14B8A6")

plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")

plt.show()
```

## Train Final K-Means Model

Based on the elbow curve, we select the optimal number of clusters and train the K-Means model.

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

clusters = kmeans.fit_predict(rfm_scaled_df)

rfm_pd["Cluster"] = clusters

rfm_pd.head()
```

```python
segment_map = {
    0: "High Value Customers",
    1: "Frequent Buyers",
    2: "At Risk Customers",
    3: "Low Value Customers"
}

rfm_pd["Segment"] = rfm_pd["Cluster"].map(segment_map)

rfm_pd.head()
```

## 📏 Cluster Evaluation Using Silhouette Score

To evaluate the quality of clustering, we compute the **Silhouette Score**, which measures how well each observation fits within its cluster.

Silhouette Score ranges from **-1 to 1**:

- Values close to **1** indicate well-separated clusters
- Values around **0** indicate overlapping clusters
- Negative values indicate incorrect clustering

```python
score = silhouette_score(rfm_scaled_df, clusters)

print("Silhouette Score:", score)
```

## 📉 Cluster Visualization using PCA

To visualize the clusters in two dimensions, we apply **Principal Component Analysis (PCA)**.

PCA reduces the dimensionality of the dataset while preserving most of the variance in the data, making it easier to visualize clustering results.

```python
pca = PCA(n_components=2)

rfm_pca = pca.fit_transform(rfm_scaled_df)

pca_df = pd.DataFrame(
    rfm_pca,
    columns=["PC1","PC2"]
)

pca_df["Cluster"] = clusters
```

```python
plt.figure(figsize=(16,9))

sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="Cluster",
    palette="Set2"
)

plt.title("Customer Segments (PCA)")
plt.show()
```

```python
cluster_summary = rfm_pd.groupby("Cluster").agg(
    {
        "Recency":"mean",
        "Frequency":"mean",
        "Monetary":"mean"
    }
)

cluster_summary
```

## 📊 Customer Segment Interpretation

Based on the clustering results, customers can be grouped into the following segments:

| Cluster   |    Segment Description   |
|-----------|--------------------------|
| Cluster 0 | High Value Customers     |
| Cluster 1 | Frequent Buyers          |
| Cluster 2 | At-Risk Customers        |
| Cluster 3 | Low Engagement Customers |

### Key Characteristics

High Value Customers:
- Low recency (recent purchases)
- High frequency
- High monetary value

At-Risk Customers:
- High recency (long time since last purchase)
- Previously moderate spending

Low Engagement Customers:
- Low purchase frequency
- Low spending levels

```python
cluster_profile = rfm_pd.groupby("Segment")[["Recency","Frequency","Monetary"]].mean()

custom_colors = [
    "#1F3A8A",  # Navy
    "#2563EB",  # Royal Blue
    "#14B8A6",  # Teal
]

cluster_profile.plot(
    kind="bar",
    figsize=(16,9),
    color=custom_colors 
)

plt.title("Customer Segment Profiles")
plt.ylabel("Average Value")
plt.xticks(rotation=0)

plt.show()
```
